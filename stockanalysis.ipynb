```python
import pandas_datareader.data as web
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
import datetime
```

```python
start= datetime.datetime(2022,1,1)
end = datetime.datetime(2022,4,17)
tcs = web.DataReader ("TCS.NS","yahoo",start,end)
infosys = web.DataReader ("INFY.NS","yahoo",start,end)
wipro = web.DataReader ("WIPRO.NS","yahoo",start,end)                      
```

```python
tcs.head()
```

```python
infosys.head()
```

```python
wipro.head()
```

```python
tcs['High'].plot(label="tcs high price",figsize=(15,7))
infosys['High'].plot(label="infy high price",figsize=(15,7))
wipro['High'].plot(label="wipro high price",figsize=(15,7))
plt.legend()

plt.show()
```

```python
tcs["Volume"].plot(label="tcs",figsize=(17,6))
infosys["Volume"].plot(label="infy")
wipro["Volume"].plot(label="wipro")
plt.legend()

plt.show()
```

```python
#when volume increase the price dropped 
wipro["High"].plot()
```

```python
#checking actual amount traded graph for infy and wipro
infosys["total traded"]=infosys["Volume"]*infosys["Open"]
wipro["total traded"]=wipro["Volume"]*wipro["Open"]
infosys["total traded"].plot(label="infy",figsize=(15,6))
wipro["total traded"].plot(label="wipro")
plt.legend()
plt.show()
```
